````python
# start point based on Lab 2 model

import torch
import torch.nn as nn
import torch.nn.functional as F

import matplotlib.pyplot as plt # for plotting
import torch.optim as optim #for gradient descent

torch.manual_seed(1) # set the random seed

class CNNClassifier(nn.Module):
    def __init__(self):
        super(CNNClassifier, self).__init__()
        self.conv1 = nn.Conv2d(3, 5, 5)
        self.pool = nn.MaxPool2d(2, 2)
        self.conv2 = nn.Conv2d(5, 10, 5)

        self.fc1 = nn.Linear(10 * 53 * 53, 32)
        self.fc2 = nn.Linear(32, 9)

    def forward(self, img):
        x = self.pool(F.relu(self.conv1(img)))
        x = self.pool(F.relu(self.conv2(x)))
        x = x.view(-1, 10 * 53 * 53)
        x = F.relu(self.fc1(x))
        x = self.fc2(x)
        return x
````

#### My Build:

I looked at some of the most successful networks and followed their trends. After each convolution, I'll do max pooling and a ReLU activation. For complexity reasons, I'll start by trying 3 convolution layers and 2 fully-connected linear layers. I want to try to get a better accuracy compared to lab 2's 2 conv layers and 2 linear layers. I believe the model in lab 2 is not complex enough to handle the image data and thus, only achieved the highest accuracy of around 72%. I'm guessing the out_channels (aka number of filters used) is arbitrary and depends on the project at hand and how complex the project is. The more filters and hidden units, the more weights need to be trained, which increases run-time and processing power required. The consequence is that you can make the model overfit to the data. I chose appropriate values for the out_channels and hidden layers by referencing LeNet and comparing values. I wanted to reduce the dimensions of the image through convolutions of stride 2 before inserting into the linear layers. I think this is the best way to isolate only the important features needed for classification.

````python
import torch
import torch.nn as nn
import torch.nn.functional as F

import matplotlib.pyplot as plt # for plotting
import torch.optim as optim #for gradient descent

torch.manual_seed(1) # set the random seed

class myCNN(nn.Module):
  def __init__(self):
    super(myCNN, self).__init__()
    self.name = "CCTV"
    # The convolution, pooling, and linear layers go here:
    self.conv1 = nn.Conv2d(in_channels = 3, out_channels = 6, kernel_size = 5, stride = 2, padding = 1)   ## I use strides of 2 to try to reduce the dimensions of the image in each conv layer
    self.pool = nn.MaxPool2d(kernel_size = 2, stride = 2)
    self.conv2 = nn.Conv2d(6, 12, 5, 2, 1)
    self.conv3 = nn.Conv2d(12, 24, 5, 2, 1)
    self.layer1 = nn.Linear(3*3*24,20)  ## Final image dimensions are 3 width by 3 height by 24 channels. This small dimensional input allows for only the important features to remain.
    self.layer2 = nn.Linear(20,9)    

  def forward(self, x):
    #print(x.size())
    x = self.pool(F.relu(self.conv1(x)))
    x = self.pool(F.relu(self.conv2(x)))
    x = self.pool(F.relu(self.conv3(x)))

    ## Could Apply a Leaky Relu to accomodate for the disadvantage of relu: when a large gradient passes through it during the backward propagation, they often become unresponsive dead-neurons.
    #x = self.pool(F.leaky_relu(self.conv1(x)))
    #x = self.pool(F.leaky_relu(self.conv2(x)))
    #x = self.pool(F.leaky_relu(self.conv3(x)))

    #print(x.size())
    x = x.view(-1, 3*3*24)
    x = F.relu(self.layer1(x))
    x = self.layer2(x)
    #x = x.squeeze(1)
    #print(x.size())
    return x
````

#### My Training

A CrossEntropyLoss was used because they are suited to classification problems and they aim to achieve zero error by penalizing differences in probability distributions between the predicted and actual probabilities more than mean squared error. Plus, this provides a softmax activation function built in, which is required for multi-class classification. 

The SGD optimizer was used because pair with a small batch size, they act as an implicit regularization to prevent overfit.

````python
## I wrote a simple Data_Loader function to ease up the process for preparing inputs to my_train function: 
def get_loader(blank_dataset, batch_size):
  # where blank is either data, val, or test. 
  a = torch.utils.data.DataLoader(blank_dataset, batch_size = batch_size, num_workers = 1, shuffle = True)
  return a
````

````python
## Apply DataLoader:
#training = get_loader(train_dataset, batch_size = 27)
#validation = get_loader(val_dataset, batch_size = 27)

def my_train(model, train_dataset, val_dataset, batch_size=27, num_epochs=1, learn_rate = 0.001):
    ### my_train accepts ImageFolder datasets and uses an internal function to computer the dataloaders, this allows for batch_size to be consistent overall.

    ## Internal DataLoader:
    train_loader = get_loader(train_dataset, batch_size = batch_size)
    val_loader =  get_loader(val_dataset, batch_size = batch_size)

    torch.manual_seed(1000)

    criterion = nn.CrossEntropyLoss()
    #optimizer = optim.Adam(model.parameters(), lr=learn_rate, weight_decay=1e-5)  ##weight decay applied to reduce overfit
    optimizer = optim.SGD(model.parameters(), lr=0.01, momentum=0.9)


    iters, losses, train_acc, val_acc = [], [], [], []

    # training
    print ("Training Started...")
    n = 0 # the number of iterations
    for epoch in range(num_epochs):
        for imgs, labels in iter(train_loader):
            
            # Applying the GPU
            if use_cuda and torch.cuda.is_available():
              imgs = imgs.cuda()
              labels = labels.cuda()

            out = model(imgs)             # forward pass
            loss = criterion(out, labels) # compute the total loss
            loss.backward()               # backward pass (compute parameter updates)
            optimizer.step()              # make the updates for each parameter
            optimizer.zero_grad()         # a clean up step for PyTorch
            n += 1
        
        # track accuracy
        iters.append(n)
        losses.append(float(loss)/batch_size)                 # compute average loss
        train_acc.append(get_accuracy(model, train_loader))   # compute training accuracy
        val_acc.append(get_accuracy(model, val_loader))       # compute valiation accuracy
        print(epoch, train_acc[-1], val_acc[-1])

    # plotting
    plt.title("Training Curve")
    plt.plot(iters, losses, label="Train")
    plt.xlabel("Iterations")
    plt.ylabel("Loss")
    plt.show()

    plt.title("Training Curve")
    plt.plot(iters, train_acc, label="Train")
    plt.plot(iters, val_acc, label="Validation")
    plt.xlabel("Iterations")
    plt.ylabel("Training Accuracy")
    plt.legend(loc='best')
    plt.show()

    print("Final Training Accuracy: {}".format(train_acc[-1]))
    print("Final Validation Accuracy: {}".format(val_acc[-1]))

    return train_acc, val_acc
````

````python
def get_accuracy(model, data_loader):
    correct = 0
    total = 0
    for imgs, labels in data_loader:
        
        if use_cuda and torch.cuda.is_available():
          imgs = imgs.cuda()
          labels = labels.cuda()

        output = model(imgs)
        #select index with maximum prediction score
        pred = output.max(1, keepdim=True)[1]
        correct += pred.eq(labels.view_as(pred)).sum().item()
        total += imgs.shape[0]
    return correct / total
````

````python
class AlexCNN(nn.Module):
  def __init__(self):
    super(AlexCNN, self).__init__()
      ## Alexnet outputs a tensor of Nx256x6x6
      ## The resolution of the 'image' is 6 by 6, which is small already so I will not do any pooling
      ## 256 is the number of channels, which makes a large number of 256 x 6 x 6.
      ### I will attempt to reduce the number of channels by using a 1D convolution

    # Number of input channels: 256 
    #self.conv1d1 = nn.Conv1d(256, 128, 5, 1, 2)  ## padding of 2 to keep dim as 6 by 6
    #self.conv1d2 = nn.Conv1d(128, 64, 5, 1, 2)

    self.layer1 = nn.Linear(256*6*6, 20)
    self.layer2 = nn.Linear(20, 9)

  def forward(self,x):
    #print(x.size())
    #x = F.relu(self.conv1d1(x))
    #x = F.relu(self.conv1d2(x))
    #print(x.size())

    x = x.view(-1, 256*6*6)
    x = F.relu(self.layer1(x))
    x = self.layer2(x)
    return x
````

````python
def get_accuracy(model, train=False):
    if train:
        data_loader = train_feature_loader
    else:
        data_loader = val_feature_loader

    correct = 0
    total = 0
    for imgs, labels in data_loader:
        
         
        #############################################
        #To Enable GPU Usage
        if use_cuda and torch.cuda.is_available():
          imgs = imgs.cuda()
          labels = labels.cuda()
        #############################################
        
        output = model(imgs)
        
        #select index with maximum prediction score
        pred = output.max(1, keepdim=True)[1]
        correct += pred.eq(labels.view_as(pred)).sum().item()
        total += imgs.shape[0]
    return correct / total
````

````python
def alex_train(model, train_feature_loader, val_feature_loader, test_feature_loader, batch_size=32, num_epochs=1):
    #train_loader = torch.utils.data.DataLoader(data, batch_size=batch_size)
    # train_loader = torch.utils.data.DataLoader(train_data, batch_size=batch_size, 
                                           #num_workers=num_workers, shuffle=True)
    criterion = nn.CrossEntropyLoss()
    optimizer = optim.SGD(model.parameters(), lr=0.01, momentum=0.9)

    iters, losses, train_acc, val_acc = [], [], [], []

    # training
    n = 0 # the number of iterations
    start_time=time.time()
    for epoch in range(num_epochs):
        ## Mini batching is used to save memory and allow for better processing power
        mini_b=0
        mini_batch_correct = 0
        Mini_batch_total = 0
        for imgs, labels in iter(train_feature_loader):
          
            
            #############################################
            #To Enable GPU Usage
            if use_cuda and torch.cuda.is_available():
              imgs = imgs.cuda()
              labels = labels.cuda()
            #############################################

          #### ALNC is alexNet.features (AlexNet without classifier) ####
           # ALNC = alexnet.features()
            out = model(imgs)             # forward pass
            loss = criterion(out, labels) # compute the total loss
            loss.backward()               # backward pass (compute parameter updates)
            optimizer.step()              # make the updates for each parameter
            optimizer.zero_grad()         # a clean up step for PyTorch



            ##### Mini_batch Accuracy ##### We don't compute accuracy on the whole trainig set in every iteration!
            pred = out.max(1, keepdim=True)[1]
            mini_batch_correct = pred.eq(labels.view_as(pred)).sum().item()
            Mini_batch_total = imgs.shape[0]
            train_acc.append((mini_batch_correct / Mini_batch_total))
            
           ###########################

          # save the current training information
            iters.append(n)
            losses.append(float(loss)/batch_size)             # compute *average* loss
            val_acc.append(get_accuracy(model, train=False))  # compute validation accuracy
            n += 1
            mini_b += 1
            print("Iteration: ",n,'Progress: % 6.2f ' % ((epoch * len(train_feature_loader) + mini_b) / (num_epochs * len(train_feature_loader))*100),'%', "Time Elapsed: % 6.2f s " % (time.time()-start_time))


        print ("Epoch %d Finished. " % epoch ,"Time per Epoch: % 6.2f s "% ((time.time()-start_time) / (epoch +1)))


    end_time= time.time()
    # plotting
    plt.title("Training Curve")
    plt.plot(iters, losses, label="Train")
    plt.xlabel("Iterations")
    plt.ylabel("Loss")
    plt.show()

    plt.title("Training Curve")
    plt.plot(iters, train_acc, label="Training")
    plt.plot(iters, val_acc, label="Validation")    
    plt.xlabel("Iterations")
    plt.ylabel("Validation Accuracy")
    plt.legend(loc='best')
    plt.show()

    train_acc.append(get_accuracy(model, train=True))
    print("Final Training Accuracy: {}".format(train_acc[-1]))
    print("Final Validation Accuracy: {}".format(val_acc[-1]))
    print ("Total time:  % 6.2f s  Time per Epoch: % 6.2f s " % ( (end_time-start_time), ((end_time-start_time) / num_epochs) ))
    
````

````python
class AlexCNN(nn.Module):
  def __init__(self):
    super(AlexCNN, self).__init__()
      ## Alexnet outputs a tensor of Nx256x6x6
      ## The resolution of the 'image' is 6 by 6, which is small already so I will not do any pooling
      ## 256 is the number of channels, which makes a large number of 256 x 6 x 6.
      ### I will attempt to reduce the number of channels by using a 1D convolution

    # Number of input channels: 256 
    #self.conv1d1 = nn.Conv1d(256, 128, 5, 1, 2)  ## padding of 2 to keep dim as 6 by 6
    #self.conv1d2 = nn.Conv1d(128, 64, 5, 1, 2)

    self.layer1 = nn.Linear(256*6*6, 20)
    self.layer2 = nn.Linear(20, 9)

  def forward(self,x):
    #print(x.size())
    #x = F.relu(self.conv1d1(x))
    #x = F.relu(self.conv1d2(x))
    #print(x.size())

    x = x.view(-1, 256*6*6)
    x = F.relu(self.layer1(x))
    x = self.layer2(x)
    return x
````

````python
def alex_train_test(model, train_feature_loader, val_feature_loader, test_feature_loader, batch_size=32, num_epochs=1):
    #train_loader = torch.utils.data.DataLoader(data, batch_size=batch_size)
    # train_loader = torch.utils.data.DataLoader(train_data, batch_size=batch_size, 
                                           #num_workers=num_workers, shuffle=True)
    criterion = nn.CrossEntropyLoss()
    optimizer = optim.SGD(model.parameters(), lr=0.01, momentum=0.9)

    iters, losses, train_acc, test_acc = [], [], [], []

    # training
    n = 0 # the number of iterations
    start_time=time.time()
    for epoch in range(num_epochs):
        mini_b=0
        mini_batch_correct = 0
        Mini_batch_total = 0
        for imgs, labels in iter(train_feature_loader):
          
            
            #############################################
            #To Enable GPU Usage
            if use_cuda and torch.cuda.is_available():
              imgs = imgs.cuda()
              labels = labels.cuda()
            #############################################

          #### ALNC is alexNet.features (AlexNet without classifier) ####
           # ALNC = alexnet.features()
            out = model(imgs)             # forward pass
            loss = criterion(out, labels) # compute the total loss
            loss.backward()               # backward pass (compute parameter updates)
            optimizer.step()              # make the updates for each parameter
            optimizer.zero_grad()         # a clean up step for PyTorch



            ##### Mini_batch Accuracy ##### We don't compute accuracy on the whole trainig set in every iteration!
            pred = out.max(1, keepdim=True)[1]
            mini_batch_correct = pred.eq(labels.view_as(pred)).sum().item()
            Mini_batch_total = imgs.shape[0]
            train_acc.append((mini_batch_correct / Mini_batch_total))
            
           ###########################

          # save the current training information
            iters.append(n)
            losses.append(float(loss)/batch_size)             # compute *average* loss
            test_acc.append(get_accuracy(model, train=False))  # compute test accuracy
            n += 1
            mini_b += 1
            print("Iteration: ",n,'Progress: % 6.2f ' % ((epoch * len(train_feature_loader) + mini_b) / (num_epochs * len(train_feature_loader))*100),'%', "Time Elapsed: % 6.2f s " % (time.time()-start_time))


        print ("Epoch %d Finished. " % epoch ,"Time per Epoch: % 6.2f s "% ((time.time()-start_time) / (epoch +1)))


    end_time= time.time()
    # plotting
    plt.title("Training Curve")
    plt.plot(iters, losses, label="Train")
    plt.xlabel("Iterations")
    plt.ylabel("Loss")
    plt.show()

    plt.title("Training Curve")
    plt.plot(iters, train_acc, label="Training")
    plt.plot(iters, test_acc, label="test")    
    plt.xlabel("Iterations")
    plt.ylabel("test Accuracy")
    plt.legend(loc='best')
    plt.show()

    train_acc.append(get_accuracy(model, train=True))
    print("Final Training Accuracy: {}".format(train_acc[-1]))
    print("Final test Accuracy: {}".format(test_acc[-1]))
    print ("Total time:  % 6.2f s  Time per Epoch: % 6.2f s " % ( (end_time-start_time), ((end_time-start_time) / num_epochs) ))
    
````
